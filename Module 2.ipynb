```python
import numpy as np
#Making Short and Long Windows
short_window = 21
long_window = 34

signals = pd.DataFrame(index = filter_Axis.index)
signals['signal'] = 0.0

#SMA of Short Window
signals['short'] = filter_Axis['Close Price'].rolling(window = short_window, min_periods = 1, center = False).mean()

#SMa of Long Window
signals['long'] = filter_Axis['Close Price'].rolling(window = long_window, min_periods = 1, center = False).mean()

#Create Signal 
signals['signal'][short_window:] = np.where(signals['short'][short_window:] > signals['long'][short_window:], 1.0,0.0)\

#Generate Trading Orders
signals['positions'] = signals['signal'].diff()
print(signals)

#filter_Axis['Date'] = pd.to_datetime(filter_Axis['Date'])
#filter_Axis.set_index('Date', inplace = True)
#filter_Axis
```

```python
fig = plt.figure(figsize=(20,10))

#subplot and label
ax1 = fig.add_subplot(111, ylabel='Price')

#Plot close Price
filter_Axis['Close Price'].plot(ax = ax1, color = 'black', lw = 2.)

#plot short amd long 
signals[['short', 'long']].plot(ax = ax1, lw = 2.)

#plot the signals
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short[signals.positions == 1.0], '^', markersize = 20, color = 'g')
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.long[signals.positions == 1.0], 'v', markersize = 20, color = 'r')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


symbol = 'TCS'
df = pd.read_csv('TCS.csv'.format(symbol), index_col = 'Date', parse_dates = True, usecols = ['Date', 'Close Price'], na_values = 'nan')
df = df.rename(columns={'Close Price': symbol})
df.dropna(inplace=True)

# simple moving average for 14 days
sma = df.rolling(window=14).mean()

# standard deviation
rstd = df.rolling(window=14).std()

upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})

try:
    
    df = df.join(upper_band).join(lower_band)
    ax = df.plot(title = '{} Price and BB'.format(symbol))
    ax.fill_between(df.index, lower_band['lower'], upper_band['upper'], color = '#ADCCFF', alpha = '0.4')
    ax.set_xlabel('Date')
    ax.set_ylabel('SMA and BB')
    ax.grid()
    plt.show()
except ValueError:
    pass
```
