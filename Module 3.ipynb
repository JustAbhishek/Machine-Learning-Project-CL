```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gold_data = pd.read_csv('Gold.csv')
gold_data.set_index('Date', inplace = True)
gold_data_nan = gold_data.dropna()
gold_data_nan
```

```python
#regression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y = np.array(gold_data_nan["Pred"])
x = np.array(gold_data_nan["new"])
y = y.reshape(-1,1)
x = x.reshape(-1,1)

regression_model = LinearRegression()
# fitting the data
regression_model.fit(x, y)
# predicting the data
y_predicted = regression_model.predict(x)

# root mean square error
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

print('Slope:', regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root Mean Squared', rmse)
print('R2 Score:', r2)


# plotting the graph
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color = 'r')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

Infy = pd.read_csv('INFY.csv', parse_dates=True, index_col='Date')
Nifty = pd.read_csv('Nifty50.csv', parse_dates=True, index_col='Date')


monthly_price = pd.concat([Infy['Close Price'], Nifty['Close']], axis=1)
monthly_price.columns = ['INFY', 'Nifty50']

print(monthly_price.head())

monthly_return = monthly_price.pct_change(1)
clean_monthly_return = monthly_return.dropna(axis=0)
print(clean_monthly_return.head())
```

```python
X = clean_monthly_return['INFY']
y = clean_monthly_return['Nifty50']

X1 = sm.add_constant(X)

model = sm.OLS(y, X1)

result = model.fit()
print(result.summary())
```
