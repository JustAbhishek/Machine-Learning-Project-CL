```python
def Average(N):
    return (dataset['Average Price'].tail(N).sum())/N
print('Average value of N days:')
print('Last 1 week', Average(5))
print('Last 2 weeks', Average(10))
print('Last 1 month', Average(20))
print('Last 3 months', Average(60))
print('Last 6 months', Average(120))
print('Last 1 Year', Average(240))

print('Profit/Loss in N days:')
def prof_loss(N):
    difference = dataset['Close Price'].tail(N).iloc[N-1] - dataset['Close Price'].tail(N).iloc[0]
    if difference < 0:
        loss = -(difference)
        loss_perc = (loss/dataset['Close Price'].tail(N).iloc[N-1])*100
        return loss_perc
    if difference > 0:
        profit = difference
        profit_perc = (profit/dataset['Close Price'].tail(N).iloc[N-1])*100
        return profit_perc
    
print('Profit/Loss in 1 week', prof_loss(5))
print('Profit/Loss in 2 weeks', prof_loss(10))
print('Profit/Loss in 1 month', prof_loss(20))
print('Profit/Loss in 3 months', prof_loss(30))
print('Profit/Loss in 6 months', prof_loss(120))
print('Profit/Loss in 1 year', prof_loss(240))
        

    
```

```python
dataset['Day_perc_change'] = dataset['Close Price'].pct_change().fillna(0)
dataset.head()
```

## PROBLEM STATEMENT 
# 1.7

```python
if((dataset['Day_perc_change'] >= -0.5) & (dataset['Day_perc_change'] <= 0.5)).all:
    dataset['Trend'] = 'Slight or No change'
if((dataset['Day_perc_change'] >= 0.5) & (dataset['Day_perc_change'] <= 1)).all():
    dataset['Trend'] = 'Slight positive'
if((dataset['Day_perc_change'] <= -0.5) & (dataset['Day_perc_change'] >= -1)).all():
    dataset['Trend'] = 'Slight negative'
if((dataset['Day_perc_change'] >= 1) & (dataset['Day_perc_change'] <= 3)).all():
    dataset['Trend'] = 'Positive'
if((dataset['Day_perc_change'] <= -1) & (dataset['Day_perc_change'] >= -3)).all():
    dataset['Trend'] = 'Negative'
if((dataset['Day_perc_change'] >= 3) & (dataset['Day_perc_change'] <= 7)).all():
    dataset['Trend'] = 'Among top gainers'
if((dataset['Day_perc_change'] <= -3) & (dataset['Day_perc_change'] >= -7)).all():
    dataset['Trend'] = 'Among top losers'
if(dataset['Day_perc_change'] > 7).all():
    dataset['Trend'] = 'Bull run'
if(dataset['Day_perc_change'] < -7).all():
    dataset['Trend'] = 'Bear drop'
dataset.head()
```
